### Analyzing Approved category

```python
# Extracting only the approved loan applications from the combined data
df_comb_App = df_comb[df_comb.NAME_CONTRACT_STATUS == 'Approved']
df_comb_App
```

#### Univariate analysis on approved loan category

```python
combined_categorical_columns = ['NAME_CONTRACT_TYPE_x',
                       'FLAG_OWN_CAR',
                       'FLAG_OWN_REALTY',
                       'CODE_GENDER',
                       'NAME_EDUCATION_TYPE',
                       'AGE_GROUP',
                       'AMT_CATEGORY']
for i in combined_categorical_columns:
    plt.figure(figsize=(10,5))
    df_comb_App[i].value_counts(normalize=True).plot.bar()
    plt.title("Approved Status: "+ i)
```

```python
combined_numerical_columns= ['AMT_GOODS_PRICE_x',
                    'AMT_INCOME_TOTAL',
                    'YEARS_EMPLOYED',
                    'YEARS_REGISTRATION',
                    'YEARS_BIRTH',
                    'AMT_CREDIT_x',
                    'AMT_ANNUITY_x']
for i in combined_numerical_columns:
    plt.figure(figsize=(20,8))
    sns.boxplot(df_comb_App[i])
    plt.title("Approved Status: "+i)
```

#### Bivariate analysis on approved loan category

```python
# Family Status
plt.hist("NAME_FAMILY_STATUS", data = df_comb_App)
plt.title("Approved status v/s Family Status")
plt.show()

# Income Type
plt.figure(figsize=(11,5))
plt.hist("NAME_INCOME_TYPE", data = df_comb_App)
plt.title("Approved status v/s Income Type")
plt.show()

# Housing type
plt.figure(figsize=(10,8))
plt.hist( "NAME_HOUSING_TYPE", data = df_comb_App)
plt.title("Approved status v/s Housing Type")
plt.show()

# CNT_CHILDREN
plt.figure(figsize=(10,7))
plt.hist("CNT_CHILDREN", data =df_comb_App)
plt.title("Approved status v/s No. pf childrens")
plt.show()

#CNT_FAM_MEMBERS
plt.figure(figsize=(10,7))
plt.hist("CNT_FAM_MEMBERS", data = df_comb_App)
plt.title("Approved status v/s No. of Family Members")
plt.show()

# Age
plt.figure(figsize=(10,7))
plt.hist(x="AGE_GROUP", data = df_comb_App)
plt.title("Approved status v/s Age Group")
plt.show()
```

#### from the above Bivariant analysis we can conclude:



Approved status v/s Family Status: People who are married are more likely to get loan approved

Approved status v/s Income Type: People who are working are more likely to get loan approved compared to students who are least likely to get loan approved

Approved status v/s Housing type: People who own House/apartment are more likely to get loan approved then compared to rented appartments/ co-op apartment types

Approved status v/s No. of children: People with 0 children are more likely to get loan approved

Approved status v/s No. of family members: If the number of people in a family is 2 they are more likely to get loan approved.

Approved status v/s Age: People with age in between 30 -50 years are more likely to get loan approved compared to the people in 20s and 60s

### Checking highest correlation coefficents in both Target 1 and Target 0

```python
# Correlation for numberical columns for Target = 0
corr = Target_0_comb.corr()
corr = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool_))
corrdf0 = corr.unstack().reset_index()
corrdf0.columns = ['VAR1', 'VAR2', 'Correlation']
corrdf0.dropna(subset = ['Correlation'], inplace = True)
corrdf0['Correlation'] = round(corrdf0['Correlation'], 2)
# Since we see correlation as an absolute value, we are converting it into absolute value
corrdf0['Correlation_abs'] = corrdf0['Correlation'].abs()
corrdf0.sort_values(by = 'Correlation_abs', ascending = False).head(10)
```

```python
# Correlation for numberical columns for Target = 1
corr = Target_1_comb.corr()

corr = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool_))
corrdf0 = corr.unstack().reset_index()
corrdf0.columns = ['VAR1', 'VAR2', 'Correlation']
corrdf0.dropna(subset = ['Correlation'], inplace = True)
corrdf0['Correlation'] = round(corrdf0['Correlation'], 2)
# Since we see correlation as an absolute value, we are converting it into absolute value
corrdf0['Correlation_abs'] = corrdf0['Correlation'].abs()
corrdf0.sort_values(by = 'Correlation_abs', ascending = False).head(10)
```
